```python
                                                    # Build an End-to-End RAG + Agent System

# ⏱️ Duration: 60 Minutes
# 📊 Total Marks: 50
# 🔷 Question 1: Build an End-to-End RAG + Agent System (25 Marks)
# 🧩 Scenario

# You are an AI intern at an ed-tech company. Students frequently ask questions about:

#     Course policies (refunds, deadlines)
#     Lecture content (PDF notes)
#     Assignment deadlines
#     Their enrollment status

# The company wants a single intelligent assistant that:

#     Answers questions using internal documents (PDFs, FAQs)
#     Fetches student-specific data (like enrollment or progress) using tools/APIs
#     Avoids hallucination and gives reliable answers

# 💻 Task

# Design and implement a working prototype (pseudo-code or real code) for this system.

# Your solution must include:
# ✅ 1. RAG Pipeline

#     How you will ingest and preprocess documents
#     Chunking strategy (with justification)
#     Embedding + vector store choice
#     Retrieval logic
#     How context is injected into the LLM

# ✅ 2. Agent Integration

#     Design an agent that decides:
#         When to use RAG
#         When to call a tool (e.g., get_student_status(student_id))
#     Show how tools are defined and used

# ✅ 3. End-to-End Flow

# Write code or structured pseudo-code showing:
#     Input query
#     Retrieval step
#     Tool calling (if needed)
#     Final answer generation

# ✅ 4. Reliability Improvements
# Show at least 2 techniques in code/design to:

#     Reduce hallucination
#     Improve answer quality

# 🎯 Expected Outcome
# A working architecture/code that demonstrates:
    # RAG + Agent working together
    # Decision-making capability
    # Real-world practicality


import os
from dotenv import load_dotenv

load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")

# Mock LLM call using Groq API (replace with real API call)
def groq_llm(prompt: str) -> str:
    """
    Mock LLM call - replace with actual Groq API call.
    Use GROQ_API_KEY from environment variable.
    """
    # Simple keyword-based responses for demo
    if "refund" in prompt.lower():
        return "Refunds are processed within 5-7 business days."
    elif "assignment" in prompt.lower():
        return "Assignments are due every Friday at 11:59 PM."
    elif "enrollment" in prompt.lower():
        return "You can check your enrollment status via the student portal."
    elif "lecture" in prompt.lower() or "notes" in prompt.lower():
        return "Lecture notes are available in PDF format on the platform."
    else:
        return "I don't know."

documents = {
    "refund_policy": "Refunds are processed in 5-7 business days.",
    "assignment_deadlines": "Assignments are due every Friday at 11:59 PM.",
    "enrollment_status": "Student enrollment status can be checked via student portal.",
    "lecture_notes": "Lecture notes are available in PDF format on the platform."
}

def retrieve_relevant_docs(query: str, docs: dict) -> str:
    """
    Simple keyword-based retrieval for demo purposes.
    Only returns docs containing meaningful keywords from query.
    """
    query_lower = query.lower()
    context = ""
    for doc_text in docs.values():
        doc_lower = doc_text.lower()
        # Only match keywords longer than 3 letters to avoid junk words
        if any(word in doc_lower for word in query_lower.split() if len(word) > 3):
            context += doc_text + "\n"
    return context.strip()

def get_student_status(student_id: str) -> str:
    """Mock student tool API."""
    return f"Student {student_id} is enrolled in 3 courses, 2 completed, 1 in progress."

def answer_query(query: str, student_id: str = None) -> str:
    """
    Main agent workflow:
    1. Retrieve context from internal docs
    2. Call tool if query requires student-specific info
    3. Inject context into LLM prompt
    4. Return final answer
    """
    # 1. Retrieve relevant docs
    context = retrieve_relevant_docs(query, documents)

    # 2. Tool integration
    if student_id and "enrollment" in query.lower():
        tool_result = get_student_status(student_id)
        context += f"\n\nStudent Status: {tool_result}"

    # 3. Build prompt
    prompt = f"""
You are an EdTech assistant. Use the context below to answer the question.

Context:
{context}

Question:
{query}

Answer concisely and accurately. If unknown, respond "I don't know".
"""
    # 4. Call LLM
    answer = groq_llm(prompt)

    # 5. Reliability improvement: fallback if answer empty
    if not answer or answer.strip() == "":
        return "I don't know"

    return answer

queries = [
    "How long does a refund take?",
    "When is the next assignment due?",
    "How do I check my enrollment status?",
    "Where can I find lecture notes?",
    "What is the refund policy for canceled courses?"
]

for q in queries:
    print(f"=== QUERY: {q} ===")
    print(answer_query(q, student_id="1001"))
    print()
```
